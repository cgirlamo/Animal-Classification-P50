```python
import pandas as pd
import geopandas as gpd
```

```python
df = pd.read_csv("804SP_Hmm.csv") # I have the time, odba, and predicted results
gdf = gpd.read_file('804_cl.shp') # I'm just GPS and time 
df
```

```python
gdf
```

```python
df['MT Time'] = pd.to_datetime(df['GMT Time']) - pd.Timedelta(hours = 6) # Make me datetime 
gdf['MT Time'] = pd.to_datetime(gdf['GMT Time']) - pd.Timedelta(hours = 6) # make me MT Time and datetime
```

```python

gdf
```

```python
times = gdf['MT Time'].to_list() # Make the index a list 
ones = {} # Create an empty dictionary for one classification results
zeros = {} # Create an empty dictionary for zero classification results
for x in times: 
    time_1 = x - pd.Timedelta(minutes = 7.5) # 7.5 minutes before each GPS point in time
    time_2 = x + pd.Timedelta(minutes = 7.5)  # 7.5 minutes after each GSP point in time
    mask = (df['MT Time']  > time_1) & (df['MT Time'] < time_2) # filter for all accelerometer results between time_1 and time_2
    x_df = df[mask] # mask the data by the above filter
    ones = x_df.loc[x_df['Results'] == 1, 'Results'].to_list() # get a list of all results which are one in the filtered dataset
    zeros = x_df.loc[x_df['Results'] == 0, 'Results'].to_list() # get a list of all results which are zero in the filtered dataset

    if len(ones) != 0 or len(zeros) != 1: # if you aren't going to divide by zero
        try:
            gdf.loc[gdf['MT Time'] == x,'ones'] = len(ones)/x_df.shape[0] # assign the classification rate
            gdf.loc[gdf['MT Time'] == x,'zeros'] = len(zeros)/x_df.shape[0]
        except ZeroDivisionError:
            continue

```
